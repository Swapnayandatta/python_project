```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/country_wise_latest.csv')
df.head()
```

```python
df.columns
```

```python
df.head(10).plot.bar(x = 'Country/Region', y ='Confirmed' )
```

```python
df['WHO Region'].value_counts().plot.pie(autopct = '%1.1f%%')
```

```python
df.head(10).groupby('Country/Region').mean().plot(kind = 'bar')
```

```python
df.head(10).plot.bar(x ='Confirmed' , y ='Deaths' )
```

```python
plt.scatter(df['New cases'], df['New deaths'],alpha= 0.7)
plt.xlim(0,100)
plt.ylim(0, 10)
plt.xlabel('New cases')
plt.ylabel('New deaths')
plt.title('Scatter Plot Example')
plt.show()
```

```python
df['Deaths / 100 Cases'].plot.hist( bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(f'Histogram of {column_name}')
```

```python
df.head(10).plot.bar(x = 'Country/Region', y = '1 week % increase')
```
